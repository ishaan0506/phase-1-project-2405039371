````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
means = df.groupby('Outcome')[['Glucose', 'BMI']].mean()
````

````python
labels = ['Non-Diabetic (0)', 'Diabetic (1)']
glucose_means = means['Glucose'].tolist()
bmi_means = means['BMI'].tolist()

x = np.arange(len(labels))  # Position of groups on the x-axis
width = 0.35
````

````python
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, glucose_means, width, label='Glucose (mg/dL)', color='#4CAF50')
rects2 = ax.bar(x + width/2, bmi_means, width, label='BMI', color='#FF9800')
````

````python
ax.set_ylabel('Average Value', fontsize=12)
ax.set_title('Average Glucose and BMI by Diabetes Outcome', fontsize=14, pad=15)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=11)
ax.legend(fontsize=11)
````

````python
ax.bar_label(rects1, padding=3, fmt='%.1f')
ax.bar_label(rects2, padding=3, fmt='%.1f')
````

````python
fig.tight_layout()
````

````python
plt.show()
````
